```python
import os
from datetime import datetime
import pandas as pd
#list that has all orders ki files ka data 
all_data = []

# saari order se start wali files and srted order
for order_file in sorted(os.listdir(".")):
    if order_file.startswith("order_data_"):
        with open(order_file, "r") as input_file:
            rows = [row.strip().split("\t") for row in input_file]

        # list of dict
        order_table = []
        for row in rows:
            order_table.append({'order_id': row[0],
                                'driver_id': row[1],
                                'passenger_id': row[2],
                                'start_region_hash': row[3],
                                'dest_region_hash': row[4],
                                'Price': float(row[5]),
                                'Time': datetime.strptime(row[6], '%Y-%m-%d %H:%M:%S')})
        
        table_namee = order_file.split(".")[0]
        order_data = {'file_name': table_namee, 'data': order_table}
        print(f"{table_namee} table created with {len(order_table)} rows")
        all_data.append(order_data)

#first order_id from the first file
print(all_data[0]['data'][0]['order_id'])
```

```python
####DANIA WE HAVE TO FIGURE OUT TH POI_CLASS WALA COLUMN IN THE TABLE



#POI Information Table
with open("poi_data", "r") as input_file:
    # splitting into columns on the basis of\t
    rows = [row.strip().split("\t") for row in input_file]

poi_list = []

# list of dict same
for row in rows:
    poi_dict = {}
    poi_dict["region_hash"] = row[0]
    
    # all poi class in dict key value pairs
    for i in range(1, len(row)):
        poi_classess = row[i].split(":")
        poi_dict[poi_classess[0]] = int(poi_classess[1])

    poi_list.append(poi_dict)

for row in poi_list:
    print(row)
```

```python
import os
from datetime import datetime

# List that has all weather files' data
all_weather_data = []

# Loop through each file and read the data
for weather_file in sorted(os.listdir(".")):
    if weather_file.startswith("weather_data_"):
        with open(weather_file, "r") as input_file:
            rows = [row.strip().split("\t") for row in input_file]

        # same as order thing
        weather_table = []
        for row in rows:
            weather_table.append({'Time': datetime.strptime(row[0], '%Y-%m-%d %H:%M:%S'),
                                  'Weather': int(row[1]),
                                  'temperature': float(row[2]),
                                  'PM2.5': float(row[3])})

        table_name = weather_file.split(".")[0]
        weather_data = {'file_name': table_name, 'data': weather_table}
        print(f"{table_name} table created with {len(weather_table)} rows")
        all_weather_data.append(weather_data)

# first timestamp from the first file
print(all_weather_data[0]['data'][0]['Time'])

```

```python
####trying extra columns like weather waghera, might wanna include most common temperature too baad mein
# now final dataset ko in intervals and the relevant fields for predictions

# import pandas as pd

dataset = pd.read_csv('final_dataset.csv')

# iss column p intervals
dataset['Time'] = pd.to_datetime(dataset['Time'])
dataset.set_index('Time', inplace=True)

# resampling
interval_ds = dataset.resample('10T').agg({'order_id': 'count', 'driver_id': 'count',
                                           'region_hash': lambda x: x.value_counts().index[0] if len(x)>0 else 0,
                                           'region_id': lambda x: x.value_counts().index[0] if len(x)>0 else 0,
                                           'temperature': lambda x: x.value_counts().index[0] if len(x)>0 else 0,
                                           'Weather': lambda x: x.value_counts().index[0] if len(x)>0 else 0})


# rename the 'order_id' and 'driver_id' columns
interval_ds.rename(columns={'order_id': 'order_count', 'driver_id': 'Supply'}, inplace=True)

interval_ds['interval'] = interval_ds.index.strftime('%Y-%m-%d %H:%M:%S')

# gap is supply demand difference
interval_ds['Gap'] = interval_ds['order_count'] - interval_ds['Supply']

# resetting the index to make 'interval' a regular column
interval_ds.reset_index(drop=True, inplace=True)

interval_ds.to_csv('interval_data1.csv', index=False)
print("Saved to file")

```

```python
# import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

# interval dataset mein predict on gap
dataset = pd.read_csv("interval_data1.csv", usecols=["order_count", "interval", "Gap"])

dataset["interval"] = pd.to_datetime(dataset["interval"])

# making interval the index
dataset.set_index("interval", inplace=True)

# rolling window of size 9 hours and 20 minutes
window_size = "9H20T"
window = dataset.rolling(window_size).sum()

#  missing values in the window with the mean warna nulls
window.fillna(window.mean(), inplace=True)

# lists to store intervals, MSE, and predicted gaps
intervals = []
mse_list = []
y_pred_list = []

#linear regression training model on the entire dataset
X = window.iloc[:-1]
y = window.shift(-1).iloc[:-1]["Gap"].fillna(window["Gap"].mean())
model = LinearRegression()
model.fit(X, y)

# iterate over the entire dataset to make predictions for all intervals
for i in range(len(X)):
    X_test = X.iloc[i:i+1]
    y_pred = model.predict(X_test)
    y_test = window.shift(-1).iloc[i+1:i+2]["Gap"].fillna(y.mean()) # fill missing values with mean
    mse = mean_squared_error(y_test, y_pred)
    interval = window.index[i+1] # Get the interval for the predicted gap and MSE

    
    intervals.append(interval)
    mse_list.append(mse)
    y_pred_list.append(y_pred[0])

mean_absolute_error = np.mean(mse_list)

print("Mean Absolute Error:", mean_absolute_error)

#dataframe to check later
results_df = pd.DataFrame({"interval": intervals, "predicted_gap": y_pred_list, "MSE": mse_list})


results_df.to_csv('linear_reg.csv', index=False)
print("Saved to file")
```
